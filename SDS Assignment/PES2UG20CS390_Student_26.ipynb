```python
data['percentage'] = ((data['math score'] + data['reading score'] + data['writing score'])/ 600 )* 100
```

```python
data['math score'].fillna(data['math score'].mean(), inplace= True)
data['reading score'].fillna(data['math score'].mean(), inplace= True)
data['writing score'].fillna(data['math score'].mean(), inplace= True)
data['percentage'].fillna(data['math score'].mean(), inplace= True)
```

```python
grades = {
    90: "s",
    80: "a",
    70: "b",
    60: "c",
    0: "d",
}
def grade_mapping(value):
    for key, letter in grades.items():
        if value >= key:
            return letter
letter_grades = data["percentage"].map(grade_mapping)
data["Final Grade"] = pd.Categorical(letter_grades, categories=grades.values(), ordered=True)
data.head()
```

```python
data = data.sample(n=100)
data
```

```python
def stratified_sample_df(data, col, n_samples):
    n = min(n_samples, data[col].value_counts().min())
    df_ = data.groupby(col).apply(lambda x: x.sample(n))
    df_.index = df_.index.droplevel(0)
    return df_
```

```python
data1=stratified_sample_df(data,'race',100)
data1
```

```python
dj = data.sample(n=100) #for a (the number in the sample is 100)
dj['math score'].mean()
```

```python
df3=stratified_sample_df(data, "race",100)
df3['math score'].mean()
```
